```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
```

```python
# Extract
csv_file = "City.csv"
city_df = pd.read_csv(csv_file)
city_df.head()
```

```python
# Transform
city_transformed = city_df.drop(["RegionID", "SizeRank", "RegionType", "StateName", "Metro"], axis=1)

city_transformed = city_transformed.rename(columns={"RegionName":"city", "State":"state", "CountyName":"county"})

city_transformed = city_transformed.dropna()

city_transformed.head()
```

```python
# Extract
csv_file = "State.csv"
state_df = pd.read_csv(csv_file)
state_df.head()
```

```python
# Transform
state_transformed = state_df.drop(["RegionID", "SizeRank", "RegionType", "StateName"], axis=1)

state_transformed = state_transformed.rename(columns={"RegionName":"state"})

state_transformed = state_transformed.dropna()

state_transformed.head()
```
